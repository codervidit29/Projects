```python
# prompt: sum of Price Each with respect to group by of Purchase Address of city

data['Purchase Address'].str.split(',').str[1].value_counts()
data['city'] = data['Purchase Address'].str.split(',').str[1]
data.groupby('city')['Price Each'].sum()
```

```python
# prompt: order by descending order of sum of Price Each for above table

data.groupby('city')['Price Each'].sum().sort_values(ascending=False)
```

```python
# prompt: percentage share of each city in terms of Price Each with respect to total Price Each for above table and order them by descending order

(data.groupby('city')['Price Each'].sum() / data['Price Each'].sum() * 100).sort_values(ascending=False)
```

```python
# prompt: graphical representation for above table

import matplotlib.pyplot as plt
city_sales = (data.groupby('city')['Price Each'].sum() / data['Price Each'].sum() * 100).sort_values(ascending=False)
plt.figure(figsize=(10, 6))
city_sales.plot(kind='bar')
plt.xlabel('City')
plt.ylabel('Percentage of Total Sales')
plt.title('Percentage of Total Sales by City')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# prompt: pie chart representation of above graph

import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
city_sales.plot(kind='pie', autopct='%1.1f%%')
plt.title('Percentage of Total Sales by City')
plt.ylabel('')  # Remove the default y-label
plt.tight_layout()
plt.show()
```

```python
# prompt: change datatype of Quantity Ordered from object to numeric

import pandas as pd
data['Quantity Ordered'] = pd.to_numeric(data['Quantity Ordered'], errors='coerce')
```

```python
data.groupby('Product')['Quantity Ordered'].sum()
```

```python
# prompt: total count of Quantity Ordered from above table

data.groupby('Product')['Quantity Ordered'].sum().sum()
```

```python
#percentage share of each product in terms of Quantity Ordered with respect to total Quantity Ordered for above table and order them by descending order

(data.groupby('Product')['Quantity Ordered'].sum() / data['Quantity Ordered'].sum() * 100).sort_values(ascending=False)
```

```python
# prompt: line graph representation for above table

import matplotlib.pyplot as plt
product_sales_percentage = (data.groupby('Product')['Quantity Ordered'].sum() / data['Quantity Ordered'].sum() * 100).sort_values(ascending=False)

plt.figure(figsize=(12, 6))
product_sales_percentage.plot(kind='line', marker='o')
plt.xlabel('Product')
plt.ylabel('Percentage of Total Quantity Ordered')
plt.title('Percentage Share of Each Product in Terms of Quantity Ordered')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
# prompt: pie chart representation for above table

import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
product_sales_percentage.plot(kind='pie', autopct='%1.1f%%')
plt.title('Percentage Share of Each Product in Terms of Quantity Ordered')
plt.ylabel('')  # Remove the default y-label
plt.tight_layout()
plt.show()
```
